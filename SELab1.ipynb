```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    # Hardcoded coefficients
    a = 0.1
    b = -1
    c = 30

    temperature_hardcoded = quadratic_model(time_values, a, b, c)

    # Include the coefficients in the plot label
    label = f' Hardcoded Coefficients: a={a}, b={b}, c={c}'
    plt.plot(time_values, temperature_hardcoded, label=label)

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Hardcoded Coefficient)')
    plt.show()

if __name__ == "__main__":
    main()
```

**Combination Of Hard Coded Values And User Choice**

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    # Multiple sets of hardcoded coefficients
    coefficients = [
        {'a': 1, 'b': -2, 'c': 5},
        {'a': -0.5, 'b': 3, 'c': 0},
        {'a': 2, 'b': -1, 'c': 2}
    ]

    # User input coefficients
    a_user = float(input("Enter the value of a: "))
    b_user = float(input("Enter the value of b: "))
    c_user = float(input("Enter the value of c: "))

    # Calculate and plot temperatures using hardcoded coefficients
    for coeff in coefficients:
        temperature = quadratic_model(time_values, coeff['a'], coeff['b'], coeff['c'])
        label = f'a={coeff["a"]}, b={coeff["b"]}, c={coeff["c"]}'
        plt.plot(time_values, temperature, label=label)

    # Calculate and plot temperatures using user input coefficients
    temperature_user = quadratic_model(time_values, a_user, b_user, c_user)
    user_label = f'User: a={a_user}, b={b_user}, c={c_user}'
    plt.plot(time_values, temperature_user, label=user_label, linestyle='--')

    # Plot the points for each value (both hardcoded and user input)
    for coeff in coefficients:
        temperature = quadratic_model(time_values, coeff['a'], coeff['b'], coeff['c'])


    # Plot the points for user input values
    plt.scatter(time_values, temperature_user, label=f'User Points: a={a_user}, b={b_user}, c={c_user}', s=10)

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Combination of user input and hardcoded Weather Modelling with Quadratic Equation')
    plt.show()

if __name__ == "__main__":
    main()

```
